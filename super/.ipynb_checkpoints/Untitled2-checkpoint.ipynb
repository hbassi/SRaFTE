```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fine_path   = "/pscratch/sd/h/hbassi/NavierStokes_fine_256_nu0.0001_k7.5.npy"
coarse_path = "/pscratch/sd/h/hbassi/NavierStokes_coarse_64_nu0.0001_k7.5.npy"
```

```python
fine   = np.load(fine_path,   mmap_mode='r')
coarse = np.load(coarse_path, mmap_mode='r')
```

```python
batch_idx = 123        # ‚Üê pick the example you care about
fine_sample   = fine[batch_idx]     # shape (T, Hf, Wf)
coarse_sample = coarse[batch_idx]   # shape (T, Hc, Wc)
T = fine_sample.shape[0]
```

```python
T
```

```python
from scipy.ndimage import zoom   # only if you want to upsample coarse ‚Üí fine

upsample = zoom   # alias so you can switch it off easily
factor   = fine_sample.shape[-1] // coarse_sample.shape[-1]   # 256//64 = 4

for t in range(0, T, 100):
    fig, axs = plt.subplots(1, 2, figsize=(8, 4), constrained_layout=True)

    # -- coarse, upsampled for apples‚Äëto‚Äëapples viewing --
    coarse_frame = coarse_sample[t]#upsample(coarse_sample[t], factor, order=1)  # bilinear
    im0 = axs[0].imshow(coarse_frame, origin="lower", cmap="viridis")
    axs[0].set_title(f"Coarse (t={t})")

    # -- fine --
    im1 = axs[1].imshow(fine_sample[t],   origin="lower", cmap="viridis")
    axs[1].set_title(f"Fine (t={t})")

    plt.pause(0.1)   # or write frames to disk if you prefer
    plt.close(fig)
```
