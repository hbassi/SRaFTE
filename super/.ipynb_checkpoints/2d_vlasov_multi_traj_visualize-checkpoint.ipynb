```python
import numpy as np
from matplotlib import pyplot as plt
import pickle
```

```python
filepath = '/pscratch/sd/e/erikaye/vm-tests/test0'
total_cg_data = []
total_fg_data = []
for i in range(10):
    if i == 6:
        continue
    cg_data_path = filepath + str(i) + '/64/'
    fg_data_path = filepath + str(i) + '/128/'

    # Load coarse-grained data
    with open(cg_data_path + "test_fe_data_64.pkl", "rb") as f:
        cg_data = pickle.load(f)
    
    # Load fine-grained data
    with open(fg_data_path + "test_fe_data_128.pkl", "rb") as f:
        fg_data = pickle.load(f)
    
    #print(i, "CG shape:", cg_data[-1].shape, "FG shape:", fg_data[-1].shape)
    
    # Create a side-by-side plot
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Coarse-grained plot
    im1 = axes[0].imshow(cg_data[-1][-1])
    axes[0].set_title(f'CG Data (i={i})')
    fig.colorbar(im1, ax=axes[0])

    # Fine-grained plot
    im2 = axes[1].imshow(fg_data[-1][-1])
    axes[1].set_title(f'FG Data (i={i})')
    fig.colorbar(im2, ax=axes[1])
    #plt.savefig(f'./figures/2d_vlasov_model_multi_traj_dataset_final_time_idx={i}.png')
    plt.show()
    total_cg_data.append(cg_data[-1])
    total_fg_data.append(fg_data[-1])
```

```python
total_cg_data = np.array(total_cg_data)
total_fg_data = np.array(total_fg_data)
total_cg_data.shape, total_fg_data.shape
```

```python
import numpy as np

def augment_sequence(input_seq, output_seq):
    """
    Apply augmentation to a full sequence.
    
    Parameters:
      input_seq: np.array, shape (81, H, W) e.g. (81, 192, 192)
      output_seq: np.array, shape (81, H', W') e.g. (81, 384, 384)
    
    Returns:
      aug_inputs: np.array, shape (num_aug, 81, H, W)
      aug_outputs: np.array, shape (num_aug, 81, H', W')
    """
    aug_inputs = []
    aug_outputs = []
    
    # Original sequence
    aug_inputs.append(input_seq)
    aug_outputs.append(output_seq)
    
    # Horizontal flip (np.fliplr applies to each 2D image in the sequence)
    aug_inputs.append(np.fliplr(input_seq))
    aug_outputs.append(np.fliplr(output_seq))
    
    # Vertical flip (np.flipud applies to each image)
    aug_inputs.append(np.flipud(input_seq))
    aug_outputs.append(np.flipud(output_seq))
    
    # 90° rotation: rotate along the spatial dimensions (axes 1 and 2)
    aug_inputs.append(np.rot90(input_seq, k=1, axes=(1,2)))
    aug_outputs.append(np.rot90(output_seq, k=1, axes=(1,2)))
    
    # 180° rotation
    aug_inputs.append(np.rot90(input_seq, k=2, axes=(1,2)))
    aug_outputs.append(np.rot90(output_seq, k=2, axes=(1,2)))
    
    # 270° rotation
    aug_inputs.append(np.rot90(input_seq, k=3, axes=(1,2)))
    aug_outputs.append(np.rot90(output_seq, k=3, axes=(1,2)))
    
    return np.array(aug_inputs), np.array(aug_outputs)

# Assuming your original data arrays have the following shapes:
# total_cg_data: (9, 81, 192, 192)
# total_fg_data: (9, 81, 384, 384)

augmented_cg_list = []
augmented_fg_list = []
num_initial_conditions = total_cg_data.shape[0]

for i in range(num_initial_conditions):
    # Get the full sequence for the i-th initial condition.
    cg_seq = total_cg_data[i]   # shape: (81, 192, 192)
    fg_seq = total_fg_data[i]   # shape: (81, 384, 384)
    
    # Apply the augmentation on the full sequence.
    aug_cg_seq, aug_fg_seq = augment_sequence(cg_seq, fg_seq)
    # Each of these has shape (6, 81, H, W), where 6 is the number of augmentations (including the original)
    
    # Append the augmented sequences.
    augmented_cg_list.append(aug_cg_seq)  # Each element is (6, 81, 192, 192)
    augmented_fg_list.append(aug_fg_seq)  # Each element is (6, 81, 384, 384)

# Now, augmented_cg_list and augmented_fg_list are lists of length 9 (initial conditions),
# each containing an array of shape (6, 81, H, W).

# Stack along a new axis to get shape (9, 6, 81, H, W)
augmented_cg = np.array(augmented_cg_list)
augmented_fg = np.array(augmented_fg_list)

# Option 1: Keep the initial condition axis separate:
#   augmented_cg.shape: (9, 6, 81, 192, 192)
#   augmented_fg.shape: (9, 6, 81, 384, 384)

# Option 2: Merge the initial condition and augmentation axes to get shape (9*6, 81, H, W)
augmented_cg = augmented_cg.reshape(-1, total_cg_data.shape[1], total_cg_data.shape[2], total_cg_data.shape[3])
augmented_fg = augmented_fg.reshape(-1, total_fg_data.shape[1], total_fg_data.shape[2], total_fg_data.shape[3])

# Now:
# augmented_cg.shape is (54, 81, 192, 192)
# augmented_fg.shape is (54, 81, 384, 384)
```
