```python
DT       = 0.010
NX = NV  = 32
#X_BOUNDS = (-math.pi,  math.pi)
#V_BOUNDS = (-6.0,      6.0)
TE_ORDER = 4                     # RK4 coarse integrator
N_CH     = 101                   # history length (channels)
```

```python
# ================================================================
# Coordinate → Fourier features
# ================================================================
class FourierFeatureMapping(nn.Module):
    def __init__(self, input_dim=2, mapping_size=64, scale=10.0):
        super().__init__()
        self.register_buffer('B', torch.randn(input_dim, mapping_size) * scale)

    def forward(self, coords):                               # (B,H,W,2)
        proj = 2 * math.pi * torch.matmul(coords, self.B)    # (B,H,W,mapping_size)
        ff   = torch.cat([torch.cos(proj), torch.sin(proj)], dim=-1)
        return ff.permute(0, 3, 1, 2)                        # (B,2*mapping_size,H,W)

# ---------------------------------------------------------------
def get_coord_grid(batch, h, w, device):
    xs = torch.linspace(0, 1, w, device=device)
    ys = torch.linspace(0, 1, h, device=device)
    gy, gx = torch.meshgrid(ys, xs, indexing='ij')
    grid = torch.stack((gx, gy), dim=-1).unsqueeze(0).repeat(batch, 1, 1, 1)
    return grid                                              # (B,H,W,2)

# ================================================================
# Fourier Neural Operator 2-D spectral layer
# ================================================================
class FourierLayer(nn.Module):
    def __init__(self, in_ch, out_ch, modes1, modes2):
        super().__init__()
        self.modes1, self.modes2 = modes1, modes2
        self.weight = nn.Parameter(
            torch.randn(in_ch, out_ch, modes1, modes2, dtype=torch.cfloat)
            / (in_ch * out_ch)
        )

    @staticmethod
    def compl_mul2d(inp, w):                                 # (B,IC,H,W) × (IC,OC,H,W)
        return torch.einsum('bixy,ioxy->boxy', inp, w)

    def forward(self, x):                                    # (B,C,H,W)  real
        B, _, H, W = x.shape
        x_ft = torch.fft.rfft2(x)

        m1 = min(self.modes1, H)
        m2 = min(self.modes2, x_ft.size(-1))                 # W_freq = W//2+1

        out_ft = torch.zeros(
            B, self.weight.size(1), H, x_ft.size(-1),
            dtype=torch.cfloat, device=x.device
        )
        out_ft[:, :, :m1, :m2] = self.compl_mul2d(
            x_ft[:, :, :m1, :m2],
            self.weight[:, :, :m1, :m2]
        )
        return torch.fft.irfft2(out_ft, s=x.shape[-2:])

# ---------------------------------------------------------------
class ConvBlock(nn.Module):
    """[Conv → GELU] × 2 (keeps H×W)."""
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.GELU(),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.GELU()
        )
    def forward(self, x): return self.block(x)

# ================================================================
# ↓↓↓ PixelShuffle-based up-sample block ↓↓↓
# ================================================================
class PixelShuffleUpsample(nn.Module):
    def __init__(self, in_ch, out_ch, upscale=2):
        super().__init__()
        self.conv = nn.Conv2d(in_ch, out_ch * (upscale ** 2), 3, padding=1)
        self.pix  = nn.PixelShuffle(upscale)
        self.act  = nn.GELU()
    def forward(self, x):
        return self.act(self.pix(self.conv(x)))

# ================================================================
# U-Net with Fourier bottleneck + PixelShuffle up-sampling
# NOTE: -- No pooling inside the bottleneck (only in encoder) --
# ================================================================
class SuperResUNet(nn.Module):
    def __init__(
        self,
        in_channels=101,
        lift_dim=128,
        mapping_size=64,
        mapping_scale=5.0,
        final_scale=2        # ← auto-detected from data
    ):
        super().__init__()

        # -------- lift ---------------
        self.fourier_mapping = FourierFeatureMapping(2, mapping_size, mapping_scale)
        lifted_ch = in_channels + 2 * mapping_size
        self.lift = nn.Conv2d(lifted_ch, lift_dim, kernel_size=1)

        # -------- encoder ------------
        self.enc1 = ConvBlock(lift_dim,        lift_dim)        # keep  (Hc)
        self.enc2 = ConvBlock(lift_dim,        lift_dim * 2)    # pool → (Hc/2)
        self.pool = nn.MaxPool2d(2)

        # -------- bottleneck ---------
        # ⚠️ Removed extra pooling here
        self.bottleneck = nn.Sequential(
            ConvBlock(lift_dim * 2, lift_dim * 2),
            FourierLayer(lift_dim * 2, lift_dim * 2, modes1=32, modes2=32),
            nn.GELU()
        )

        # -------- decoder ------------
        # up1 keeps spatial dims (upscale=1) so it matches e2
        self.up1  = PixelShuffleUpsample(lift_dim * 2, lift_dim * 2, upscale=1)
        self.dec2 = ConvBlock(lift_dim * 4, lift_dim)                    # cat(up1,e2)

        self.up2  = PixelShuffleUpsample(lift_dim, lift_dim)             # ×2  (Hc/2 → Hc)
        self.dec1 = ConvBlock(lift_dim * 2, lift_dim // 2)               # cat(up2,e1)

        self.dec0 = nn.Sequential(                                       # Hc → Hc×final_scale
            PixelShuffleUpsample(lift_dim // 2, lift_dim // 4, upscale=final_scale),
            ConvBlock(lift_dim // 4, lift_dim // 4)
        )

        # -------- output head --------
        self.out_head = nn.Sequential(
            nn.Conv2d(lift_dim // 4, 32, 3, padding=1),
            nn.GELU(),
            nn.Conv2d(32, in_channels, 3, padding=1)                      # 11-channel output
        )

    # -----------------------------------------------------------
    def forward(self, x):                                 # (B,11,Hc,Wc) normalised
        B, _, H, W = x.shape
        coords = get_coord_grid(B, H, W, x.device)
        x = torch.cat([x, self.fourier_mapping(coords)], dim=1)   # lift
        x = self.lift(x)

        e1 = self.enc1(x)               # Hc
        e2 = self.enc2(self.pool(e1))   # Hc/2

        # ---- bottleneck (no extra pooling) ----
        b  = self.bottleneck(e2)        # Hc/2

        # ---- decoder ----
        d2 = self.up1(b)                             # Hc/2  (spatially matches e2)
        d2 = self.dec2(torch.cat([d2, e2], dim=1))

        d1 = self.up2(d2)                            # Hc
        d1 = self.dec1(torch.cat([d1, e1], dim=1))

        d0 = self.dec0(d1)                           # Hf
        return self.out_head(d0)                     # (B,11,Hf,Wf)  normalised
```

```python
sys.path.append(os.path.abspath("vlasov_fd"))
from setup_.configs import *
from setup_.enums import BCType, FDType
from axis import Axis
from coord.coord_sys import Coordinate, CoordinateType
from coord.cartesian import CartesianCoordinateSpace
from grid import Grid
from field import ScalarField
from pde_vlasovES import VlasovPoisson
sys.argv = _saved_argv

class VlasovStepper:
    def __init__(self):
        # Integration parameters
        self.dt       = DT
        self.te_order = TE_ORDER
        # Species parameter configurations
        
        plasma_config = UnitsConfiguration()
        charge = plasma_config.e  # unit of charge
        eV = plasma_config.eV     # unit of kB*T
        permittivity = plasma_config.eps0
        n0_e = 1.0             # electron number density
        n0_i = n0_e            # ion number density
        T_e  = 1.0             # electron temperature [eV]
        T_i  = T_e             # ion temperature [eV]
        
        mass_ratio = 25.
        mass_e = 1.0          # mass of electron
        mass_i = mass_ratio*mass_e    # mass of ion
        
        self.ion_config = IonConfiguration(n0=n0_i, mass=mass_i, T=T_i, units_config=plasma_config)
        self.elc_config = ElcConfiguration(n0=n0_e, mass=mass_e, T=T_e, units_config=plasma_config)
        vth_e = self.elc_config.vth    # electron thermal speed
        wp_e  = self.elc_config.wp     # plasma e' frequency
        lamD  = self.elc_config.lamD   # Debye length
        wp_i  = self.ion_config.wp     # plasma i+ frequency
        vth_i = self.ion_config.vth    # ion thermal speed
        k = 0.10/lamD
        v0 = wp_e/k     ## avg flow of electrons with respect to ions, k set s.t. v0 >> vth_e
        
        X_BOUNDS = -np.pi/k, np.pi/k
        VE_BOUNDS = -6 * v0, 6 * v0
        VI_BOUNDS = -50 * vth_i, 50 * vth_i

        # Coordinates and axes
        X = Coordinate('X', CoordinateType.X)
        V = Coordinate('V', CoordinateType.X)
        x_pts = np.linspace(X_BOUNDS[0], X_BOUNDS[1], NX, endpoint=False)
        ve_pts = np.linspace(VE_BOUNDS[0], VE_BOUNDS[1], NV, endpoint=False)
        vi_pts = np.linspace(VI_BOUNDS[0], VI_BOUNDS[1], NV, endpoint=False)
        self.ax_x  = Axis(NX, coordinate=X, xpts=x_pts)
        self.ax_ve = Axis(NV, coordinate=V, xpts=ve_pts)
        self.ax_vi = Axis(NV, coordinate=V, xpts=vi_pts)

        # Coordinate spaces
        self.coords_x  = CartesianCoordinateSpace('X',  self.ax_x)
        self.coords_ve = CartesianCoordinateSpace('Ve', self.ax_ve)
        self.coords_vi = CartesianCoordinateSpace('Vi', self.ax_vi)

        # Grids
        self.grid_X = Grid('X',  (self.ax_x,))
        self.grid_e = Grid('XVe',(self.ax_x, self.ax_ve))
        self.grid_i = Grid('XVi',(self.ax_x, self.ax_vi))

        # Derivative configurations
        self.deriv_x  = DerivativeConfiguration(left_bc=BCType.PERIODIC,
                                                order=self.te_order,
                                                fd_type=FDType.CENTER)
        self.deriv_ve = DerivativeConfiguration(left_bc=BCType.ZEROGRADIENT,
                                                order=self.te_order,
                                                fd_type=FDType.CENTER)
        self.deriv_vi = DerivativeConfiguration(left_bc=BCType.ZEROGRADIENT,
                                                order=self.te_order,
                                                fd_type=FDType.CENTER)

        # Laplacian MPO for Poisson solve
        self.lapl_mpo = self.grid_X.laplacian_mpo(
            ax_deriv_configs={self.ax_x: self.deriv_x}
        )

        # Uniform BC for potential
        self.bc = (np.ones(self.ax_x.npts) / self.ax_x.dx, None)

      
        # self.elc_config = ElcConfiguration()
        # self.ion_config = IonConfiguration()

    def step_block(self, ce, ci):
        """
        Advance every slice in the (B, T, NX, NV) tensors by one timestep
        using a flattened batch and tqdm progress bar, while suppressing
        all solver prints.

        Args:
            ce (Tensor): shape (B, T, NX, NV)
            ci (Tensor): shape (B, T, NX, NV)

        Returns:
            Tensor: ce_pred of shape (B, T, NX, NV)
        """
        B, T, NX, NV = ce.shape
        BT = B * T
        
        # Flatten time and batch dims
        ce_flat = ce.view(BT, NX, NV)
        ci_flat = ci.view(BT, NX, NV)
        pred_flat = torch.zeros((BT, NX, NV), dtype=ce.dtype, device=ce.device)

        # Loop with progress bar, suppressing solver prints inside
        for m in trange(BT, desc="Vlasov stepping", leave=False):
            with open(os.devnull, 'w') as fnull, \
                 contextlib.redirect_stdout(fnull), \
                 contextlib.redirect_stderr(fnull):
                fe_np = ce_flat[m].cpu().numpy().ravel()
                fi_np = ci_flat[m].cpu().numpy().ravel()

                # Build MPS/GTN
                fe_gtn = self.grid_e.make_gridTN(fe_np)
                fe_gtn.ax_deriv_configs = {self.ax_x: self.deriv_x,
                                           self.ax_ve: self.deriv_ve}
                fi_gtn = self.grid_i.make_gridTN(fi_np)
                fi_gtn.ax_deriv_configs = {self.ax_x: self.deriv_x,
                                           self.ax_vi: self.deriv_vi}

                # Compute charge density
                charge_e = fe_gtn.integrate(integ_axes=[self.ax_ve],
                                            is_sqrt=False,
                                            new_grid=self.grid_X)
                charge_e.scalar_multiply(-1.0, inplace=True)
                charge_i = fi_gtn.integrate(integ_axes=[self.ax_vi],
                                            is_sqrt=False,
                                            new_grid=self.grid_X)
                charge_density = charge_i.add(charge_e)
                charge_density.ax_deriv_configs = {self.ax_x: self.deriv_x}

                # Poisson solve
                V_gtn   = charge_density.solve(self.lapl_mpo)
                V_field = ScalarField('V', self.grid_X, data=V_gtn, is_sqrt=False)

                # Wrap fields and step
                fe_field = ScalarField('fe', self.grid_e, data=fe_gtn, is_sqrt=False)
                fi_field = ScalarField('fi', self.grid_i, data=fi_gtn, is_sqrt=False)
                vp_sys = VlasovPoisson(
                    fe_field, fi_field, V_field,
                    coords_x=self.coords_x,
                    coords_ve=self.coords_ve,
                    coords_vi=self.coords_vi,
                    elc_params=self.elc_config,
                    ion_params=self.ion_config,
                    upwind=False,
                    evolve_ion=False,
                    potential_boundary_conditions=self.bc,
                    te_order=self.te_order
                )
                vp_sys = vp_sys.next_time_step(
                    self.dt,
                    compress_level=1,
                    is_first_time_step=(m < B)
                )

                # Extract and store
                new_fe = vp_sys.fe.get_comp_data()
                pred_flat[m] = torch.from_numpy(new_fe).float().to(ce.device)

        # Reshape back to (B, T, NX, NV)
        ce_pred = pred_flat.view(B, T, NX, NV)
        return ce_pred
```

```python
def evaluate_and_plot(sample_idx):
    from IPython import get_ipython
    get_ipython().run_line_magic('matplotlib', 'inline')
    import numpy as np
    import matplotlib.pyplot as plt
    from IPython.display import clear_output
    import torch.nn.functional as F

    # constants for charge density
    qe = -1.0
    dv = {32: 0.375, 128: 0.09375}

    # extract full sequences
    ce_seq = coarse_e[sample_idx]  # (T, Hc, Wc)
    ci_seq = coarse_i[sample_idx]
    fe_seq = fine_e[sample_idx]    # (T, Hf, Wf)

    # initialize sliding windows
    ce_win = ce_seq[:N_CH].to(device)
    ci_win = ci_seq[:N_CH].to(device)

    stepper = VlasovStepper()
    num_steps = fe_seq.shape[0] - N_CH
    Nf = fe_seq.shape[-1]  # fine grid size

    for step in trange(num_steps, desc=f"Eval sample {sample_idx}"):
        # prepare batch dims
        ce_in = ce_win.unsqueeze(0)
        ci_in = ci_win.unsqueeze(0)

        # coarse propagation
        with torch.no_grad():
            ce_pred = stepper.step_block(ce_in, ci_in).squeeze(0)

        # update windows
        ce_win = torch.cat([ce_win[1:], ce_pred[-1].detach().unsqueeze(0)], dim=0)
        next_ci = ci_seq[N_CH + step].to(device).unsqueeze(0)
        ci_win = torch.cat([ci_win[1:], next_ci], dim=0)

        # super-resolution (un-normalize)
        x_norm = (ce_pred.unsqueeze(0) - data_mean) / data_std
        with torch.no_grad():
            fine_pred = model(x_norm).squeeze(0)
        fine_pred = fine_pred * data_std + data_mean

        # get predicted & true fine slices
        pred_f = fine_pred[-1].cpu().numpy()       # (Hf, Wf)
        true_f = fe_seq[N_CH + step].cpu().numpy() # (Hf, Wf)

        # compute charge density for predicted and true
        rho_pred = qe * pred_f.sum(axis=-1) * dv[Nf]
        rho_true = qe * true_f.sum(axis=-1) * dv[Nf]

        # compute baseline from coarse_e: upsample coarse_e slice to (Hf, Wf)
        coarse_slice = ce_seq[N_CH + step].unsqueeze(0).unsqueeze(0).to(device)  # (1,1,Hc,Wc)
        coarse_up = F.interpolate(coarse_slice, size=(Nf, Nf), mode='bicubic', align_corners=False)
        coarse_up = coarse_up.squeeze(0).squeeze(0).cpu().numpy()
        rho_base = qe * coarse_up.sum(axis=-1) * dv[Nf]

        # errors
        mae_pred = np.mean(np.abs(rho_pred - rho_true))
        mae_base = np.mean(np.abs(rho_base - rho_true))
        print(f"Step {step:3d}: MAE_pred = {mae_pred:.3e}, MAE_base = {mae_base:.3e}")

        # clear and plot
        clear_output(wait=True)
        plt.figure(figsize=(6,4))
        plt.plot(rho_true, label='Ground Truth')
        plt.plot(rho_pred, '--', label='Prediction')
        plt.plot(rho_base, ':', label='Coarse Baseline')
        plt.title(f"Step {step} Charge Density\nMAE_pred={mae_pred:.3e}, MAE_base={mae_base:.3e}")
        plt.xlabel('Spatial index')
        plt.ylabel('ρ(x)')
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.close()
```

```python
import numpy as np
from scipy.stats import wasserstein_distance
import matplotlib.pyplot as plt

# ─── Single‐step metric functions ────────────────────────────────────────────

def wasserstein_step(true_f, pred_f, base_f=None):
    """
    Compute Wasserstein‐1 distance for a single frame.
    Inputs are 2D arrays (H, W).
    Returns (w_pred, w_base) or (w_pred, None).
    """
    v = true_f.ravel()
    # shift so nonnegative
    vmin = v.min()
    v_true = v - vmin
    v_pred = pred_f.ravel() - vmin
    w_pred = wasserstein_distance(v_true, v_pred)
    if base_f is None:
        return w_pred, None
    v_base = base_f.ravel() - vmin
    w_base = wasserstein_distance(v_true, v_base)
    return w_pred, w_base

def relative_L2_step(true_f, pred_f, base_f=None, eps=1e-8):
    """
    Compute relative L2 error ||pred-true||₂/||true||₂ for one frame.
    """
    v_true = true_f.ravel()
    norm   = np.linalg.norm(v_true) + eps
    err_pred = np.linalg.norm(pred_f.ravel() - v_true) / norm
    if base_f is None:
        return err_pred, None
    err_base = np.linalg.norm(base_f.ravel() - v_true) / norm
    return err_pred, err_base

def KL_step(true_f, pred_f, base_f=None, eps=1e-10):
    """
    Compute D_KL(true||pred) for one frame; normalizes each to sum=1.
    """
    p = np.abs(true_f)
    p /= p.sum() + eps
    q_pred = np.abs(pred_f)
    q_pred /= q_pred.sum() + eps
    kl_pred = np.sum(p.ravel() * np.log((p.ravel() + eps)/(q_pred.ravel() + eps)))
    if base_f is None:
        return kl_pred, None
    q_base = np.abs(base_f)
    q_base /= q_base.sum() + eps
    kl_base = np.sum(p.ravel() * np.log((p.ravel() + eps)/(q_base.ravel() + eps)))
    return kl_pred, kl_base

# ─── Plotting helper ─────────────────────────────────────────────────────────

def plot_metric_series(metric_dict, title, ylabel):
    for name, arr in metric_dict.items():
        plt.plot(arr, label=name)
    plt.xlabel('time step')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def evaluate_and_plot(sample_idx):
    from IPython import get_ipython
    get_ipython().run_line_magic('matplotlib', 'inline')
    import numpy as np
    import matplotlib.pyplot as plt
    from IPython.display import clear_output
    import torch.nn.functional as F

    qe = -1.0
    dv = {32: 0.375, 128: 0.09375}

    ce_seq = coarse_e[sample_idx]               # (T, Hc, Wc)
    ci_seq = fine_i[sample_idx][:, ::4, ::4]    # (T, Hc, Wc)
    fe_seq = fine_e[sample_idx]                 # (T, Hf, Wf)

    ce_win = ce_seq[:N_CH].to(device)
    ci_win = ci_seq[:N_CH].to(device)

    stepper   = VlasovStepper()
    num_steps = fe_seq.shape[0] - N_CH
    Nf        = fe_seq.shape[-1]
    ws_pred, ws_base = [], []
    r2_pred, r2_base = [], []
    kl_pred, kl_base = [], []

    for step in trange(num_steps, desc=f"Eval sample {sample_idx}"):
        # propagate
        ce_in   = ce_win.unsqueeze(0)
        ci_in   = ci_win.unsqueeze(0)
        with torch.no_grad():
            ce_pred = stepper.step_block(ce_in, ci_in).squeeze(0)

        ce_win = torch.cat([ce_win[1:], ce_pred[-1].unsqueeze(0)], dim=0)
        next_ci = ci_seq[N_CH + step].to(device).unsqueeze(0)
        ci_win  = torch.cat([ci_win[1:], next_ci], dim=0)

        # super-res
        x_norm = (ce_pred.unsqueeze(0) - data_mean) / data_std
        with torch.no_grad():
            fine_pred = model(x_norm).squeeze(0)
        fine_pred = fine_pred * data_std + data_mean

        pred_f = fine_pred[-1].cpu().numpy()
        true_f = fe_seq[N_CH + step].cpu().numpy()
        mae_pred = np.mean(np.abs(pred_f - true_f))

        # build the bicubic upsample baseline
        coarse_slice = ce_seq[N_CH + step].unsqueeze(0).unsqueeze(0).to(device)
        coarse_up = F.interpolate(coarse_slice, size=(Nf, Nf),
                                  mode='bicubic', align_corners=False)
        coarse_up = coarse_up.squeeze().cpu().numpy()
        base_f = coarse_up
    
        # compute and store each metric
        w_p, w_b   = wasserstein_step(true_f, pred_f, base_f)
        r2_p, r2_b = relative_L2_step(true_f, pred_f, base_f)
        kl_p, kl_b = KL_step(true_f,   pred_f, base_f)
    
        ws_pred.append(w_p); ws_base.append(w_b)
        r2_pred.append(r2_p); r2_base.append(r2_b)
        kl_pred.append(kl_p); kl_base.append(kl_b)
    
        # optional: print current metrics
        print(f"Step {step:3d}: W_pred={w_p:.3e}, W_base={w_b:.3e}, "
              f"r2_pred={r2_p:.3e}, r2_base={r2_b:.3e}, "
              f"KL_pred={kl_p:.3e}, KL_base={kl_b:.3e}")
        mae_base = np.mean(np.abs(coarse_up - true_f))
        print(f"Step {step:3d}: MAE_base = {mae_base:.3e}, MAE_pred = {mae_pred:.3e}")
        # Build a multi‐line metrics string
        metrics_str = (
            f"Wₚ={w_p:.2e}, W_b={w_b:.2e}\n"
            f"r2ₚ={r2_p:.2e}, r2_b={r2_b:.2e}\n"
            f"KLₚ={kl_p:.2e}, KL_b={kl_b:.2e}\n"
            f"MAEₚ={mae_pred:.2e}, MAE_b={mae_base:.2e}"
        )

        clear_output(wait=True)
        fig, axes = plt.subplots(1, 3, figsize=(12, 4), constrained_layout=True)
    
        # get global vmin/vmax
        vmin = min(true_f.min(), base_f.min(), pred_f.min())
        vmax = max(true_f.max(), base_f.max(), pred_f.max())
    
        # 1) Ground truth
        im0 = axes[0].imshow(true_f, aspect='auto', vmin=vmin, vmax=vmax)
        axes[0].set_title("Ground Truth")
        axes[0].axis('off')
    
        # 2) Upsampled baseline
        im1 = axes[1].imshow(base_f, aspect='auto', vmin=vmin, vmax=vmax)
        axes[1].set_title("Bicubic Upsample")
        axes[1].axis('off')
        # metrics string for baseline
        txt_b = (
            f"MAE={mae_base:.2e}\n"
            f"W₁={w_b:.2e}\n"
            f"r₂={r2_b:.2e}\n"
            f"KL={kl_b:.2e}"
        )
        # place in upper-left corner inside the axes
        axes[1].text(
            0.02, 0.98, txt_b,
            transform=axes[1].transAxes,
            va='top', ha='left',
            color='white', fontsize=9,
            bbox=dict(facecolor='black', alpha=0.5, pad=2)
        )
    
        # 3) Neural prediction
        im2 = axes[2].imshow(pred_f, aspect='auto', vmin=vmin, vmax=vmax)
        axes[2].set_title("Predicted")
        axes[2].axis('off')
        # metrics string for prediction
        txt_p = (
            f"MAE={mae_pred:.2e}\n"
            f"W₁={w_p:.2e}\n"
            f"r₂={r2_p:.2e}\n"
            f"KL={kl_p:.2e}"
        )
        axes[2].text(
            0.02, 0.98, txt_p,
            transform=axes[2].transAxes,
            va='top', ha='left',
            color='white', fontsize=9,
            bbox=dict(facecolor='black', alpha=0.5, pad=2)
        )
    
        # single colorbar
        cbar = fig.colorbar(im2, ax=axes, location='right', shrink=0.8)
        cbar.ax.set_ylabel('f-value', rotation=270, labelpad=15)
    
        plt.show()
        plt.close(fig)
```
