```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib. pyplot as plt
from tqdm import trange
import math
```

```python
# ================================================================
# Coordinate → Fourier features
# ================================================================
class FourierFeatureMapping(nn.Module):
    def __init__(self, input_dim=2, mapping_size=64, scale=10.0):
        super().__init__()
        self.register_buffer('B', torch.randn(input_dim, mapping_size) * scale)

    def forward(self, coords):                               # (B,H,W,2)
        proj = 2 * math.pi * torch.matmul(coords, self.B)    # (B,H,W,mapping_size)
        ff   = torch.cat([torch.cos(proj), torch.sin(proj)], dim=-1)
        return ff.permute(0, 3, 1, 2)                        # (B,2*mapping_size,H,W)

# ---------------------------------------------------------------
def get_coord_grid(batch, h, w, device):
    xs = torch.linspace(0, 1, w, device=device)
    ys = torch.linspace(0, 1, h, device=device)
    gy, gx = torch.meshgrid(ys, xs, indexing='ij')
    grid = torch.stack((gx, gy), dim=-1).unsqueeze(0).repeat(batch, 1, 1, 1)
    return grid                                              # (B,H,W,2)

# ================================================================
# Fourier Neural Operator 2-D spectral layer
# ================================================================
class FourierLayer(nn.Module):
    def __init__(self, in_ch, out_ch, modes1, modes2):
        super().__init__()
        self.modes1, self.modes2 = modes1, modes2
        self.weight = nn.Parameter(
            torch.randn(in_ch, out_ch, modes1, modes2, dtype=torch.cfloat)
            / (in_ch * out_ch)
        )

    @staticmethod
    def compl_mul2d(inp, w):                                 # (B,IC,H,W) × (IC,OC,H,W)
        return torch.einsum('bixy,ioxy->boxy', inp, w)

    def forward(self, x):                                    # (B,C,H,W)  real
        B, _, H, W = x.shape
        x_ft = torch.fft.rfft2(x)

        m1 = min(self.modes1, H)
        m2 = min(self.modes2, x_ft.size(-1))                 # W_freq = W//2+1

        out_ft = torch.zeros(
            B, self.weight.size(1), H, x_ft.size(-1),
            dtype=torch.cfloat, device=x.device
        )
        out_ft[:, :, :m1, :m2] = self.compl_mul2d(
            x_ft[:, :, :m1, :m2],
            self.weight[:, :, :m1, :m2]
        )
        return torch.fft.irfft2(out_ft, s=x.shape[-2:])

# ---------------------------------------------------------------
class ConvBlock(nn.Module):
    """[Conv → GELU] × 2 (keeps H×W)."""
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.GELU(),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.GELU()
        )
    def forward(self, x): return self.block(x)

# ================================================================
# ↓↓↓ PixelShuffle-based up-sample block ↓↓↓
# ================================================================
class PixelShuffleUpsample(nn.Module):
    def __init__(self, in_ch, out_ch, upscale=2):
        super().__init__()
        self.conv = nn.Conv2d(in_ch, out_ch * (upscale ** 2), 3, padding=1)
        self.pix  = nn.PixelShuffle(upscale)
        self.act  = nn.GELU()
    def forward(self, x):
        return self.act(self.pix(self.conv(x)))

# ================================================================
# U-Net with Fourier bottleneck + PixelShuffle up-sampling
# NOTE: -- No pooling inside the bottleneck (only in encoder) --
# ================================================================
class SuperResUNet(nn.Module):
    def __init__(
        self,
        in_channels=101,
        lift_dim=128,
        mapping_size=64,
        mapping_scale=5.0,
        final_scale=2        # ← auto-detected from data
    ):
        super().__init__()

        # -------- lift ---------------
        self.fourier_mapping = FourierFeatureMapping(2, mapping_size, mapping_scale)
        lifted_ch = in_channels + 2 * mapping_size
        self.lift = nn.Conv2d(lifted_ch, lift_dim, kernel_size=1)

        # -------- encoder ------------
        self.enc1 = ConvBlock(lift_dim,        lift_dim)        # keep  (Hc)
        self.enc2 = ConvBlock(lift_dim,        lift_dim * 2)    # pool → (Hc/2)
        self.pool = nn.MaxPool2d(2)

        # -------- bottleneck ---------
    
        self.bottleneck = nn.Sequential(
            ConvBlock(lift_dim * 2, lift_dim * 2),
            FourierLayer(lift_dim * 2, lift_dim * 2, modes1=32, modes2=32),
            nn.GELU()
        )

        # -------- decoder ------------
        # up1 keeps spatial dims (upscale=1) so it matches e2
        self.up1  = PixelShuffleUpsample(lift_dim * 2, lift_dim * 2, upscale=1)
        self.dec2 = ConvBlock(lift_dim * 4, lift_dim)                    # cat(up1,e2)

        self.up2  = PixelShuffleUpsample(lift_dim, lift_dim)             # ×2  (Hc/2 → Hc)
        self.dec1 = ConvBlock(lift_dim * 2, lift_dim // 2)               # cat(up2,e1)

        self.dec0 = nn.Sequential(                                       # Hc → Hc×final_scale
            PixelShuffleUpsample(lift_dim // 2, lift_dim // 4, upscale=final_scale),
            ConvBlock(lift_dim // 4, lift_dim // 4)
        )

        # -------- output head --------
        self.out_head = nn.Sequential(
            nn.Conv2d(lift_dim // 4, 32, 3, padding=1),
            nn.GELU(),
            nn.Conv2d(32, in_channels, 3, padding=1)                      # 11-channel output
        )

    # -----------------------------------------------------------
    def forward(self, x):                                 # (B,11,Hc,Wc) normalised
        B, _, H, W = x.shape
        coords = get_coord_grid(B, H, W, x.device)
        x = torch.cat([x, self.fourier_mapping(coords)], dim=1)   # lift
        x = self.lift(x)

        e1 = self.enc1(x)               # Hc
        e2 = self.enc2(self.pool(e1))   # Hc/2

        # ---- bottleneck ----
        b  = self.bottleneck(e2)        # Hc/2

        # ---- decoder ----
        d2 = self.up1(b)                             # Hc/2  (spatially matches e2)
        d2 = self.dec2(torch.cat([d2, e2], dim=1))

        d1 = self.up2(d2)                            # Hc
        d1 = self.dec1(torch.cat([d1, e1], dim=1))

        d0 = self.dec0(d1)                           # Hf
        return self.out_head(d0)                     # (B,11,Hf,Wf)  normalised
```

```python
from matplotlib import gridspec
from scipy.ndimage import zoom
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------------------------------
# helper: LaTeX-style scientific notation (no $ wrapper here—added in title)
def tex_sci(x, prec=2):
    s = f"{x:.{prec}e}"
    base, exp = s.split('e')
    return rf"{base} \times 10^{{{int(exp):d}}}"
# -----------------------------------------------------------------

eps       = 1e-8            # numerical stability
case_list = [10]

for case in case_list:
    for t in range(0, Nt, 10):

        # ---------- data & up-sampling -------------------------------
        rho_cg   = input_data[case, t]
        rho_up   = zoom(rho_cg, scale, order=3)
        rho_pred = pred[case, t].cpu().detach().numpy()
        rho_fine = output_data[case, t].cpu().detach().numpy()

        # ---------- colour limits -----------------------------------
        vmin, vmax = rho_fine.min(), rho_fine.max()

        # ---------- error maps & scale ------------------------------
        err_up   = np.abs(rho_up   - rho_fine)
        err_pred = np.abs(rho_pred - rho_fine)
        err_max  = max(err_up.max(), err_pred.max())

        # ---------- relative L2 errors ------------------------------
        rl2_up   = np.linalg.norm(rho_up   - rho_fine) / (np.linalg.norm(rho_fine) + eps)
        rl2_pred = np.linalg.norm(rho_pred - rho_fine) / (np.linalg.norm(rho_fine) + eps)

        # ---------- figure layout -----------------------------------
        fig = plt.figure(figsize=(12, 6))
        gs  = gridspec.GridSpec(
            2, 4,
            width_ratios=[1, 1, 1, 0.05],
            wspace=0.05,
            hspace=0.25          # vertical gap between rows
        )

        # axes -------------------------------------------------------
        ax00 = fig.add_subplot(gs[0, 0]); ax01 = fig.add_subplot(gs[0, 1]); ax02 = fig.add_subplot(gs[0, 2])
        ax10 = fig.add_subplot(gs[1, 0]); ax11 = fig.add_subplot(gs[1, 1]); ax12 = fig.add_subplot(gs[1, 2])

        # ---------- top row: fields ---------------------------------
        field_panels = [rho_cg, rho_up, rho_pred]
        field_titles = [
            "True coarse dynamics",
            "Upsampled coarse dynamics",
            "Predicted fine dynamics"
        ]

        for ax, img, ttl in zip([ax00, ax01, ax02], field_panels, field_titles):
            h_field = ax.imshow(img, origin="lower", vmin=vmin, vmax=vmax)
            ax.set_title(ttl, fontsize=14, fontweight='bold')
            ax.set_xlabel(r"$x$", fontsize=12, fontweight='bold', labelpad=-7)   # closer x-label
            ax.set_ylabel(r"$y$", fontsize=12, fontweight='bold', labelpad=-7)   # closer v-label
            ax.set_xticks([0, img.shape[1]-1]); ax.set_xticklabels([0, img.shape[1]], fontsize=10)
            ax.set_yticks([0, img.shape[0]-1]); ax.set_yticklabels([img.shape[0], 0], fontsize=10)
            ax.grid(color='white', linewidth=0.5, linestyle='--', alpha=0.3)

        # ---------- bottom row: fine truth & errors ------------------
        error_panels = [rho_fine, err_up, err_pred]
        error_titles = [
            "True fine dynamics",
            rf"Upsampled (error = ${tex_sci(rl2_up)}$)",
            rf"FUnet (error = ${tex_sci(rl2_pred)}$)"
        ]

        for ax, img, ttl in zip([ax10, ax11, ax12], error_panels, error_titles):
            if ttl.startswith("True"):
                h_err = ax.imshow(img, origin="lower", vmin=vmin, vmax=vmax)
            else:
                h_err = ax.imshow(img, origin="lower", cmap="jet", vmin=0, vmax=err_max)
            ax.set_title(ttl, fontsize=12, fontweight='bold')
            ax.set_xlabel(r"$x$", fontsize=12, fontweight='bold', labelpad=-7)   # closer x-label
            ax.set_ylabel(r"$y$", fontsize=12, fontweight='bold', labelpad=-7)   # closer v-label
            ax.set_xticks([0, img.shape[1]-1]); ax.set_xticklabels([0, img.shape[1]], fontsize=10)
            ax.set_yticks([0, img.shape[0]-1]); ax.set_yticklabels([img.shape[0], 0], fontsize=10)
            ax.grid(color='white', linewidth=0.5, linestyle='--', alpha=0.3)

        # ---------- colour-bars -------------------------------------
        cax_field = fig.add_subplot(gs[0, 3])
        cax_err   = fig.add_subplot(gs[1, 3])

        fig.colorbar(h_field, cax=cax_field, orientation="vertical", label=r"$|\Psi|$")
        fig.colorbar(h_err,   cax=cax_err,   orientation="vertical", label="error")

        # ---------- save / show -------------------------------------
        fig.savefig(
            f'./figures/tdse_results/funet/'
            f'funet_tdse_phase1_test_fe_ic={case}_t={t}.pdf',
            bbox_inches="tight", dpi=300
        )
        plt.show()
```

```python
# ---------- user choices ---------------------------------------
import os
case_list_md = [10]          # trajectories to analyse
stride_md    = 10            # time stride (matches earlier code)
out_dir_md   = './figures/tdse_results/funet/marginals/'
os.makedirs(out_dir_md, exist_ok=True)

eps = 1e-12
err_up_all, err_pred_all = [], []

# choose up to 5 frames, spaced by 2·stride_md = 20 steps
times_to_plot = list(range(0, Nt))[:5]   # 0,20,40,60,80

# ---------- main loops -----------------------------------------
for case in case_list_md:

    # ------------------------------------------------------------
    #   1 × 5 figure with marginal ρx at five times
    # ------------------------------------------------------------
    fig, axs = plt.subplots(1, 5, figsize=(15, 3), sharey=True,
                            gridspec_kw={'wspace': 0.25})

    for idx, t in enumerate(times_to_plot):

        # --- amplitudes (|Ψ|) -----------------------------------
        amp_cg   = input_data[case, t]                       # coarse amplitude
        amp_up   = zoom(amp_cg, scale, order=3)              # bicubic up-sample
        amp_pred = pred[case, t].cpu().numpy()               # network output
        amp_true = output_data[case, t].cpu().numpy()

        # --- probability densities -----------------------------
        prob_up, prob_pred, prob_true = amp_up**2, amp_pred**2, amp_true**2

        # --- marginal ρx ---------------------------------------
        rho_x_up   = np.sum(prob_up,   axis=0)
        rho_x_pred = np.sum(prob_pred, axis=0)
        rho_x_true = np.sum(prob_true, axis=0)

        # --- grids ---------------------------------------------
        Ny, Nx = prob_true.shape
        xgrid  = np.linspace(La, Lb, Nx)

        # ---------- plotting -----------------------------------
        ax = axs[idx]
        ax.plot(xgrid, rho_x_true, lw=3, c='k',        label="Ground truth" if idx==0 else "")
        ax.plot(xgrid, rho_x_up,   lw=3, c='tab:blue', ls='--', label="Upsampled"    if idx==0 else "")
        ax.plot(xgrid, rho_x_pred, lw=3, c='tab:red',  ls='--', label="FUnet"        if idx==0 else "")

        ax.set_title(f"t = {round(t*0.01,1)}", fontweight='bold')
        ax.set_xlabel("x", fontweight='bold')
        if idx == 0:
            ax.set_ylabel(r"$\rho_x$", fontweight='bold')
            ax.legend(fontsize=8, frameon=False)

        # ---------- per‑snapshot rel‑L2 errors ------------------
        vec_up    = rho_x_up
        vec_pred  = rho_x_pred
        vec_truth = rho_x_true

        rl2_up   = np.linalg.norm(vec_up   - vec_truth)/(np.linalg.norm(vec_truth)+eps)
        rl2_pred = np.linalg.norm(vec_pred - vec_truth)/(np.linalg.norm(vec_truth)+eps)

        err_up_all.append(rl2_up)
        err_pred_all.append(rl2_pred)

    # ---------- save & close whole 1×5 figure -------------------
    fig.suptitle("Marginal ρx predictions", fontsize=14, fontweight='bold')
    fig.tight_layout(rect=[0, 0, 1, 0.92])
    fig.savefig(f"{out_dir_md}/marginal_tdse_ic={case}.pdf",
                dpi=300, bbox_inches="tight")
    plt.close(fig)

# ---------- summary print‑out ----------------------------------
print("\n--- Marginal densities: mean rel‑L2 over all snapshots ---")
print(f"Upsampled  : {np.mean(err_up_all):.4e}")
print(f"Prediction : {np.mean(err_pred_all):.4e}\n")
```

```python
# ---------- constants already defined in your script -----------
tunits = 2.4188843265857e-17            # 1 a.u.  in seconds
dt_au  = dt                             # dt from your earlier code
dt_fs  = dt_au * tunits * 1e15          # one integration step in femtoseconds

# ---------- user choices ---------------------------------------
case_list_dp = [10]
stride_dp    = 10                       # same stride you use to sample frames
out_dir_dp   = './figures/tdse_results/funet/dipoles/'
os.makedirs(out_dir_dp, exist_ok=True)

eps = 1e-12
err_up_all, err_pred_all = [], []

for case in case_list_dp:

    dip_x_true, dip_y_true = [], []
    dip_x_up,   dip_y_up   = [], []
    dip_x_pred, dip_y_pred = [], []
    time_fs                = []        # collects physical times

    for step in range(0, Nt, stride_dp):

        # amplitudes |Ψ|
        amp_cg   = input_data[case, step]
        amp_up   = zoom(amp_cg, scale, order=3)
        amp_pred = pred[case, step].cpu().detach().numpy()
        amp_true = output_data[case, step].cpu().detach().numpy()

        prob_up, prob_pred, prob_true = amp_up**2, amp_pred**2, amp_true**2

        Ny, Nx = prob_true.shape
        h      = (Lb - La)/(Nx - 1)
        xgrid  = np.linspace(La, Lb, Nx)
        ygrid  = np.linspace(La, Lb, Ny)
        X, Y   = np.meshgrid(xgrid, ygrid, indexing='xy')

        d_true = np.array([np.sum(X * prob_true),  np.sum(Y * prob_true)]) * h**2
        d_up   = np.array([np.sum(X * prob_up),    np.sum(Y * prob_up)])   * h**2
        d_pred = np.array([np.sum(X * prob_pred),  np.sum(Y * prob_pred)]) * h**2

        dip_x_true.append(d_true[0]); dip_y_true.append(d_true[1])
        dip_x_up.append(d_up[0]);     dip_y_up.append(d_up[1])
        dip_x_pred.append(d_pred[0]); dip_y_pred.append(d_pred[1])

        # per-snapshot rel-L2 error
        rl2_up   = np.linalg.norm(d_up   - d_true)/(np.linalg.norm(d_true)+eps)
        rl2_pred = np.linalg.norm(d_pred - d_true)/(np.linalg.norm(d_true)+eps)
        err_up_all.append(rl2_up)
        err_pred_all.append(rl2_pred)

        # physical time (fs)
        time_fs.append(step * dt_fs)

    # ---------- plot full time series ---------------------------
    fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10, 4))

    ax0.plot(time_fs, dip_x_true, lw=2,  c='k',        label='True fine')
    ax0.plot(time_fs, dip_x_up,   lw=1.3,c='tab:blue', label='Upsampled')
    ax0.plot(time_fs, dip_x_pred, lw=1.3,c='tab:red',  label='Prediction')
    ax0.set_title(r"Dipole $d_x(t)$", fontweight='bold')
    ax0.set_xlabel(r"Time [fs]", fontweight='bold');  ax0.set_ylabel(r"$d_x$", fontweight='bold')
    ax0.legend()

    ax1.plot(time_fs, dip_y_true, lw=2,  c='k',        label='True fine')
    ax1.plot(time_fs, dip_y_up,   lw=1.3,c='tab:blue', label='Upsampled')
    ax1.plot(time_fs, dip_y_pred, lw=1.3,c='tab:red',  label='Prediction')
    ax1.set_title(r"Dipole $d_y(t)$", fontweight='bold')
    ax1.set_xlabel(r"Time [fs]", fontweight='bold');  ax1.set_ylabel(r"$d_y$", fontweight='bold')
    ax1.legend()

    fig.tight_layout()
    # fig.savefig(
    #     f"{out_dir_dp}/dipole_tdse_ic={case}.pdf",
    #     dpi=300, bbox_inches="tight")
    plt.show()
    plt.close(fig)

# ---------- summary print-out ----------------------------------
print("\n--- Dipole moment: mean rel-L2 over all snapshots ---")
print(f"Upsampled  : {np.mean(err_up_all):.4e}")
print(f"Prediction : {np.mean(err_pred_all):.4e}\n")
```

```python
N_cases = input_data.shape[0]
N_t     = input_data.shape[1]

# Arrays to store per-case mean **relative ℓ₂** errors
case_rel_pr = np.zeros(N_cases)
case_rel_up = np.zeros(N_cases)

for case in trange(N_cases, desc="cases"):
    rel_pr_list = []
    rel_up_list = []

    for t in range(N_t):
        # -------- phase-space slice and bicubic upsample ----------
        f_cg   = input_data[case, t].cpu().numpy()
        f_up   = zoom(f_cg, Nf / Nc, order=3)
        f_pred = mean_pred[case, t].cpu().numpy()
        f_fine = output_data[case, t].cpu().numpy()

        # -------- grid-wise relative ℓ₂ error ---------------------
        rel_pr_list.append(
            np.linalg.norm(f_pred - f_fine) / (np.linalg.norm(f_fine) + eps)
        )
        rel_up_list.append(
            np.linalg.norm(f_up - f_fine) / (np.linalg.norm(f_fine) + eps)
        )

    # average over timesteps for this case
    case_rel_pr[case] = np.mean(rel_pr_list)
    case_rel_up[case] = np.mean(rel_up_list)

# ------------------ summary across cases ----------------------
print(f"\nMean relative ℓ₂ error across the full test set (N = {N_cases} cases):")
print(f"  Model     : mean = {case_rel_pr.mean():.4e},  std = {case_rel_pr.std():.4e}")
print(f"  Bicubic   : mean = {case_rel_up.mean():.4e},  std = {case_rel_up.std():.4e}")
```

```python
def radial_energy_spectrum(u):
    """One-sided, Parseval-normalised radial spectrum."""
    N      = u.shape[0]
    u_hat  = np.fft.fftshift(np.fft.fft2(u))
    energy = np.abs(u_hat)**2 / N**2
    kx     = np.fft.fftshift(np.fft.fftfreq(N)) * N
    KX, KY = np.meshgrid(kx, kx, indexing='ij')
    kr     = np.sqrt(KX**2 + KY**2).astype(int)
    E      = np.bincount(kr.ravel(), energy.ravel(),
                         minlength=kr.max()+1)
    return np.arange(len(E)), E
```

```python
from scipy.ndimage import zoom                 # already imported earlier
case_list = [8]
dt       = 0.01
Nyquist  = rho_fine.shape[0] // 4
scale    = rho_fine.shape[0] // input_data.shape[-1]   # e.g. 256/128 = 2
eps      = 1e-12                                      # avoid divide-by-0
pad = lambda arr, L: np.pad(arr, (0, L-len(arr)), 'constant')

for case in case_list:

    # --- containers for time-averaging ---------------------------------
    spec_gt_ts, spec_pred_ts, spec_up_ts = [], [], []
    k_ref = None                     # reference k-axis (max length seen)

    for t in range(0, Nt, 10):

        # ---------- fetch fields ---------------------------------------
        rho_cg   = input_data [case, t]
        rho_up   = zoom(rho_cg, scale, order=3)
        rho_pred = mean_pred  [case, t].cpu().detach().numpy()
        rho_fine = output_data[case, t].cpu().detach().numpy()

        # ---------- radial spectra -------------------------------------
        k_gt,  E_gt   = radial_energy_spectrum(rho_fine)
        _,     E_pred = radial_energy_spectrum(rho_pred)
        _,     E_up   = radial_energy_spectrum(rho_up)

        # ---------- keep the longest k-axis seen so far -----------------
        L = len(k_gt)
        if k_ref is None or L > len(k_ref):
            k_ref = np.arange(L)

        spec_gt_ts  .append(pad(E_gt  , len(k_ref)))
        spec_pred_ts.append(pad(E_pred, len(k_ref)))
        spec_up_ts  .append(pad(E_up  , len(k_ref)))

        # ---------- Nyquist line index for this grid -------------------
        i_nyq = Nyquist
        k_nyq = k_ref[i_nyq]

        # ---------- snapshot plot --------------------------------------
        plt.figure(figsize=(6, 4))
        plt.loglog(k_ref, pad(E_gt,   len(k_ref)), 'k-',   lw=2.0, label='Ground truth')
        plt.loglog(k_ref, pad(E_pred, len(k_ref)), 'C3--', lw=2.0, label='FUnet')
        plt.loglog(k_ref, pad(E_up,   len(k_ref)), 'C1:',  lw=2.0, label='Bicubic upsample')
        #plt.axvline(k_nyq, color='0.5', lw=1.0)
        plt.xlabel(r'radial wavenumber $k$')
        plt.ylabel(r'$E(k)$')
        plt.title(f'Case {case},  t = {t*dt:.2f}')
        plt.grid(True, which='both', ls='--', alpha=0.3)
        plt.legend()
        plt.show()

    # ===================================================================
    #  FINAL TIME-AVERAGED PLOT FOR THIS CASE
    # ===================================================================
    E_gt_avg   = np.mean(np.stack(spec_gt_ts  , axis=0), axis=0)
    E_pred_avg = np.mean(np.stack(spec_pred_ts, axis=0), axis=0)
    E_up_avg   = np.mean(np.stack(spec_up_ts  , axis=0), axis=0)

    plt.figure(figsize=(6, 4))
    plt.loglog(k_ref, E_gt_avg,   'k-',   lw=2.4, label='Ground truth')
    plt.loglog(k_ref, E_pred_avg, 'C3--', lw=2.2, label='FUnet')
    plt.loglog(k_ref, E_up_avg,   'C1:',  lw=2.2, label='Bicubic upsample')
    #plt.axvline(k_ref[Nyquist], color='0.5', lw=1.0)
    plt.xlabel(r'radial wavenumber $k$')
    plt.ylabel(r'$E(k)$')
    plt.title(f'Time-averaged radial energy spectra')
    plt.grid(True, which='both', ls='--', alpha=0.3)
    plt.legend()
    plt.savefig(f'./figures/tdse_results/funet/radial_energy_spectrum_test_case{case}_time_avg.pdf')
    plt.show()
```
